# Reshaping,Stacking,Squeezing,Unsqueezing and permute

*   Reshape - reshaping a tensor into a certain shape - tensor.reshape(x,y,z)
*   View - Return a view of an input tensor of certain shape but  keep thhe same memory as the origina tensor
* Stacking - combine multiple tensors on top of each other(vstack) or side by side(hstatck)
* Squeeze   - removes all `1` dimensions from a tensor
* Unsqueeze add a `1` dimesion to a tensor
* Premute -  Return a view of the input wiht dimensiones permuted(swaped) in a certain way

```python
 import torch
x=torch.arange(1,20,2)
x.shape
print(x)
print("x:reshape",x.reshape(5,2))
print("x after reshape:",x)
#VIEW
z=x.view(2,5)#z is given as a view now z shares x's memory
print("z:\n",z)
z[: ,1:2]=99
print("x after changing element in z:\n",x)#whatever change made in z will be made in x
#STACK
x_vstacked=torch.stack([x,x,x,x],dim=0)
x_hstacked=torch.stack([x,x,x,x],dim=1)
print("x_vstacked:\n",x_vstacked)
print("x_hstacked:\n",x_hstacked)
#SQUEEZE
tensor=torch.rand(1,5)
print("tensor:\n",tensor)
print("tensor.shape:",tensor.shape)
tensor_squeezed=tensor.squeeze()
print("tensor_squeezed:\n",tensor_squeezed)
#UNSQUEEZED
tensor_unsqueezed=tensor_squeezed.unsqueeze(dim=1)
print("tensor_unsqueezed\n",tensor_unsqueezed)
#PERMUTE
x=torch.rand(1,2,3)
print("x shape:",x.shape)
x_permuted=x.permute(1,2,0)
print("x permuted shpe:",x_permuted.shape)
print("x:\n",x)
print("_permuted\n",x_permuted)
```

# INDEXING
indexing in Pytorch is similar to indexing in numPy

```python
import torch
x=torch.arange(0,27).reshape(3,3,3)
print(x)
print("selecting all\n:",x[:,:,:])
print("\nselecting first row first row:\n",x[0:1,0:1,:])
print("\nSelecting first two rows of second row:\n",x[1:2,0:2,:])
print("\n selecting first term of first row of last row:\n",x[-1,0:1,0])
print("\n selecting 3rd column of all three rows of first row:\n",x[0,:,-1])
print("\n Selecting 1st column of all thress rows\n:",x[:,:,0])
print('\n Selecting 2 columns from each row\n:',x[:,:,0:2])
print("\n Selecting the middle element of each row\n:",x[:,1:2,1:2])
```

# REPRODUCIBILITY
Trying to take random out of random
* start with random numbers
* tensor operations
* update random numbers to try and make them better reperesentation of the data
* again and
* again

To reduce the randomnes in neural networks and pYtorch comes the concept of a **random seed**
*  you give r random seed intilizing
*  then attach it to `torch.manual_seed(seednumber)`
*  call this everytime u create a random tensor which u want to have same random numbers


```python
R_A=torch.rand(3,4)
R_B=torch.rand(3,4)
print(R_A==R_B)
#hence we use torch.manual_seed(random-seed)
random_seed=12
torch.manual_seed(random_seed)
R_C=torch.rand(3,4)
torch.manual_seed(random_seed)#each time u create a random tensor call the seeed
R_D=torch.rand(3,4)
print(R_C==R_D)
```
